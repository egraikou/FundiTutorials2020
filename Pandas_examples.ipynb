<font size="3"><b>Plot the number of pulsar discoveries through time</b></font>

```python
def plotyear(year):
    data = ppdotDiagram[ppdotDiagram.year <= year]
    
    dataUL = data.query('comp == "UL"')
    dataCO = data.query('comp == "CO"')
    dataNS = data.query('comp == "NS"')
    dataHe = data.query('comp == "He"')                    
        
#    dataUL = data[data.BINCOMP.str.contains('UL')]
#    dataCO = data[data.BINCOMP.str.contains('CO')]
#    dataNS = data[data.BINCOMP.str.contains('NS')]
#    dataHe = data[data.BINCOMP.str.contains('He')]
    
    ax1 = data.plot.scatter('p0','p1',
                      linewidths=3,edgecolors='k', marker ='.',
                      figsize=(20,15), label = 'Isolated pulsars')   
    if not dataUL.empty:
        ax2 = dataUL.plot.scatter('p0','p1',
                      linewidths=4,edgecolors='y', marker = '<',
                      figsize=(20,15), label = 'Ultra light companion', ax = ax1)
    if not dataCO.empty:
        ax3 = dataCO.plot.scatter('p0', 'p1',
                        linewidths=4,edgecolors='b', marker ='s',
                        figsize=(20,15), label = 'CO companion', ax = ax1)
    if not dataHe.empty:
        ax4 = dataHe.plot.scatter('p0', 'p1',
                        linewidths=4,edgecolors='r',
                        figsize=(20,15), label = 'He companion', ax = ax1)
    if not dataNS.empty:
        ax5 = dataNS.plot.scatter('p0', 'p1',
                        linewidths=5, edgecolors='green', marker = 'D', 
                        figsize=(20,15), label = 'NS companion', ax = ax1)

    plt.axis(ymin=10e-23, ymax=10e-9, xmin=0.001, xmax=30)
    plt.xscale('log')
    plt.yscale('log')
    ax1.tick_params(axis='both', which='major', labelsize=20)
    ax1.tick_params(axis="x", direction="in", length=16, width=2, color="blue")
    plt.legend(loc=2, prop={'size': 20})
    plt.xlabel('log(P0)', fontsize=25)
    plt.ylabel('log(P1)', fontsize=25)
```

```python
interact(plotyear,year=widgets.IntSlider(min=1968,max=2019,step=1,value=2000))
```

```python
data = {'year': ([]),
        'pulsar_no': ([])
        }

count = 0
for i in range(0, len(sorted_ppdot)-1):
    if sorted_ppdot.year.iloc[i]== sorted_ppdot.year.iloc[i+1]:
        count = count +1
    else:
        data['pulsar_no'].append(count+1)
        data['year'].append(sorted_ppdot.year.iloc[i])

print data['pulsar_no'], data['year']
plt.figure(figsize=(20,10))
plt.plot(data['year'], data['pulsar_no'], 'k')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Year', fontsize=25)
plt.ylabel('Cumulative number of pulsars', fontsize=25)
```

```python
gapminder[(gapminder.country == 'Greece') & (gapminder.year == 1990)]
```

```python
def plot_babies(year):
    babies = gapminder[gapminder.year == year]
    area = 5e-6 * babies.population
    colors = babies.region.map({'Africa': 'skyblue', 'Europe': 'gold', 'America': 'palegreen', 'Asia': 'coral'})
                   
        
    ax1 = babies.plot.scatter('babies_per_woman','age5_surviving',
                      c = colors, s = area,
                      linewidths=1, marker ='.', edgecolors='k',
                      figsize=(20,15))   

    plt.axis(ymin=0, ymax=110, xmin=0.0, xmax=9.0)
    plt.grid('on')
    ax1.tick_params(axis='both', which='major', labelsize=20)
    plt.xlabel('Babies per woman', fontsize=25)
    plt.ylabel('Child mortality', fontsize=25)
```

```python
interact(plot_babies,year=widgets.IntSlider(min=1950,max=2015,step=1,value=2000))
```
